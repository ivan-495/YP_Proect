1. market_file.csv содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.
    - id — номер покупателя в корпоративной базе данных.
    - Покупательская активность — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
    - Тип сервиса — уровень сервиса, например «премиум» и «стандарт». Зависит от наличия определенной подписки у пользователя.
    - Разрешить сообщать — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
    - Маркет_актив_6_мес — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
    - Маркет_актив_тек_мес — количество маркетинговых коммуникаций в текущем месяце.
    - Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
    - Акционные_покупки — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
    - Популярная_категория — самая популярная категория товаров у покупателя за последние 6 месяцев.
    - Средний_просмотр_категорий_за_визит — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.
    - Неоплаченные_продукты_штук_квартал — общее число неоплаченных товаров в корзине за последние 3 месяца.
    - Ошибка_сервиса — число сбоев, которые коснулись покупателя во время посещения сайта за последние 3 месяца.- Страниц_за_визит — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

2. market_money.csv данные о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.
    - id — номер покупателя в корпоративной базе данных.
    - Период — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'.
    - Выручка — сумма выручки за период в условных единицах.

3. market_time.csv данные о времени (в минутах), которое покупатель провёл на сайте в течение периода.
    - id — номер покупателя в корпоративной базе данных.
    - Период — название периода, во время которого зафиксировано общее время.
    - минут — значение времени, проведённого на сайте, в минутах.

4. money.csv данные о среднемесячной прибыли продавца за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.
    - id — номер покупателя в корпоративной базе данных.
    - Прибыль — значение прибыли в условных единицах.

Проведите исследовательский анализ данных из каждой таблицы. Отберите клиентов с покупательской активностью не менее трёх месяцев, то есть таких, которые что-либо покупали в этот период. Оформите выводы по результатам шага.

### Датафрейм market_file.csv

```python
market_file.head(5)
```

```python
# Исключаем столбец 'id'
numerical_cols = market_file.select_dtypes(include=['int64', 'float64']).columns.drop('id')
categorical_cols = market_file.select_dtypes(include=['object']).columns

print("Числовые столбцы:", numerical_cols)
print("Категориальные столбцы:", categorical_cols)
```

```python
#Описательная статистика для числовых столбцов
numerical_stats = market_file[numerical_cols].describe().T
print(numerical_stats)
```

```python
#Количество выбросов по правилу IQR
for col in numerical_cols:
    Q1 = market_file[col].quantile(0.25)
    Q3 = market_file[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = market_file[(market_file[col] < lower_bound) | (market_file[col] > upper_bound)]
    print(f"Количество выбросов в столбце {col}: {len(outliers)}")
```

```python
#Статистика для категориальных столбцов
categorical_stats = market_file[categorical_cols].describe(include='object').T
print(categorical_stats)
```

```python
#Визуализация для числовых столбцов
for col in numerical_cols:
    # Создаем фигуру с двумя подграфиками
    plt.figure(figsize=(12, 5))
    
    # Гистограмма
    plt.subplot(1, 2, 1)
    sns.histplot(data=market_file, x=col, kde=True)
    plt.title(f'Гистограмма для {col}')
    
    # Диаграмма размаха (boxplot)
    plt.subplot(1, 2, 2)
    sns.boxplot(data=market_file, y=col)
    plt.title(f'Boxplot для {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Список столбцов для визуализации
columns_to_plot = [
    'маркет_актив_тек_мес',
    'средний_просмотр_категорий_за_визит',
    'неоплаченные_продукты_штук_квартал',
    'ошибка_сервиса',
    'страниц_за_визит'
]

# Построение графиков для каждого столбца
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=market_file)
    plt.title(f'Частота уникальных значений {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.xticks(rotation=45)  # Поворот меток по оси X для лучшей читаемости
    plt.show()
```

```python
# Настраиваем визуализацию
plt.figure(figsize=(15, 10))
plt.suptitle('Распределение категориальных признаков', y=1.02, fontsize=14)

# Строим отдельный график для каждого категориального признака
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2, 2, i)
    
    # Сортируем по частоте встречаемости значений
    order = market_file[col].value_counts().index
    
    sns.countplot(data=market_file, 
                 x=col, 
                 order=order,
                 palette='viridis')
    
    plt.title(f'Распределение {col}')
    plt.xlabel('')
    plt.xticks(rotation=45)
    plt.ylabel('Количество наблюдений')
    
    # Добавляем аннотации с процентным соотношением
    total = len(market_file)
    for p in plt.gca().patches:
        height = p.get_height()
        plt.gca().text(p.get_x() + p.get_width()/2.,
                       height + 20,
                       f'{height/total:.1%}',
                       ha='center')

plt.tight_layout()
plt.show()
```

```python
# Создание сводной таблицы по выручке
money_pivot = market_money.pivot_table(index='id', columns='период')
money_pivot.columns = ['выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц']
money_pivot = money_pivot.assign(id=money_pivot.index)

# Создание сводной таблицы по времени
time_pivot = market_time.pivot_table(index='id', columns='период')
time_pivot.columns = ['минут_предыдущий_месяц', 'минут_текущий_месяц']
time_pivot = time_pivot.assign(id=time_pivot.index)

# Объединение данных в единый датафрейм
merged_data = (
    market_file
    .join(money_pivot.set_index('id'), on='id', rsuffix='_выручка')
    .join(time_pivot.set_index('id'), on='id', rsuffix='_минут')
    .reset_index(drop=True)
)

# Удаление лишних столбцов
result = merged_data.drop(columns=['id_выручка', 'id_минут'], errors='ignore')
```

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False))
    ]
)
```

```python
# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
ord_pipe = Pipeline(
    [
        (
            'simple_imputer_ord_before',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(categories=[
                                      ['Редко покупает по акции','Часто покупает по акции']],
                          handle_unknown='use_encoded_value',
                          unknown_value=np.nan)
        ),
        (
            'simple_imputer_ord_after',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 7),
        'models__max_features': range(2,7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='poly')],
        'models__degree': range(2, 3),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
randomized_search.fit(X_train, y_train)
```

```python
print('Лучшая модель с параметрами:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', round(randomized_search.best_score_, 2))
```

```python
# проверьте работу модели на тестовой выборке
# рассчитайте прогноз на тестовых данных
y_test_pred = randomized_search.predict(X_test)
y_test_proba = randomized_search.predict_proba(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {round(roc_auc_score(y_test, y_test_proba[:,1]), 2)}')
```

```python
def histogram(df, col, target):
    plt.figure(figsize=(8,6))
    plot = sns.histplot(df, bins=20, kde=True, hue=target, x=col)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_ylabel('Количество', fontsize=14)
    
    
def categoral_unique(df, col):
    display(df[col].unique())
    
    plt.figure(figsize=(8,6))
    plot = sns.countplot(y=col, data=df)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_xlabel('Количество', fontsize=14)
    
    
def hist_with_wiskers(df, col, target):
    sns.set()
    f, axes = plt.subplots(1, 2, figsize=(16, 4))
    axes[0].set_title(f'Гистограмма для {col}', fontsize=16)
    axes[0].set_ylabel('Количество', fontsize=14)
    if target != None:
        sns.histplot(df, bins=20, kde=True, ax=axes[0], hue=target, x=col)
    else:
        sns.histplot(df, bins=20, kde=True, ax=axes[0], x=col)
    axes[1].set_title(f'График ящик с усами для {col}', fontsize=16)
    sns.boxplot(data=df, ax=axes[1], y=col)
    axes[1].set_ylabel(col, fontsize=14)
    plt.show()

def pivot_bar_plot(df, col):
    plt.figure(figsize=(8,6))
    plot = sns.barplot(x=col, data=df, y=df.index)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_xlabel('Количество', fontsize=14)
    
def plot_discrete_distribution(df, col, target):
    plt.figure(figsize=(8, 6))
    plot = sns.countplot(data=df, x=col, hue=target)
    plot.set_title(f'Распределение по {col}', fontsize=16)
    plot.set_ylabel('Количество', fontsize=14)
    plt.xticks(rotation=45)
    plt.show()
```
